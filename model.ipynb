
<h2>What is Alzheimer's Disease</h2>
A progressive disease that destroys memory and other important mental functions.
Brain cell connections and the cells themselves degenerate and die, eventually destroying memory and other important mental functions.


<h2>About Dataset</h2>
This dataset contains extensive health information for 2,149 patients, each uniquely identified with IDs ranging from 4751 to 6900. The dataset includes demographic details, lifestyle factors, medical history, clinical measurements, cognitive and functional assessments, symptoms, and a diagnosis of Alzheimer's Disease. The data is ideal for researchers and data scientists looking to explore factors associated with Alzheimer's, develop predictive models, and conduct statistical analyses.

```python
# splitting input and output data
X,y = df.iloc[:,:-1], df.iloc[:,-1]
print(X.shape)
print(y.shape)
```

```python
X.head()
```

```python
# performing train test split on X and y
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, chi2

```

```python
# selecting 8 best features from the dataset
selector = SelectKBest(score_func=chi2,k=8)
X_train_updated = selector.fit_transform(X_train, y_train)
X_test_updated = selector.transform(X_test)
```

```python
# the 8 best features from the data set
best_feat = X.columns[selector.get_support(indices=True)]
best_feat
```

```python
# fitting the model to training data after standardizing the input data where mean will be 0 and standard deviation would be 1
from sklearn.metrics import accuracy_score
std_scaler = StandardScaler()
X_scaled = std_scaler.fit_transform(X)
rfc = RandomForestClassifier()
rfc.fit(X_train_updated, y_train)
```

```python

# prediction of the model
y_pred = rfc.predict(X_test_updated)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# applying hyperparameter tuning using GridSearchCV for getting the best values for the parameters of random forest classifier model
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators" : [100,200,250,300],
    "max_features" : ["auto","sqrt","log2"],
    "max_depth" : [4,5,6,7,8],
    
}

grid_search = GridSearchCV(RandomForestClassifier(), param_grid=param_grid, n_jobs=-1)
grid_search.fit(X_train_updated,y_train)
grid_search.best_params_
```

```python
# fitting the model after hyperparameter tuning with setting the best values of it's paramters
model = RandomForestClassifier(max_depth=8, max_features='sqrt', n_estimators=200)
model.fit(X_train_updated,y_train)
y_pred = model.predict(X_test_updated)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
